```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("owid-covid-data.csv")
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
threshold = len(df) * 0.05
threshold
```

```python
col_to_drop = df.columns[ (df.isna().sum() <= threshold) & (df.isna().sum() > 0) ]
col_to_drop
```

we can't drop any missing values because they exceed the threshold

```python
df_subset = df.drop(df[ df['location'].isin(['High income', 'Upper middle income','Lower middle income',"World","European Union"]) ].index)
```

## total number of cases in each location

```python
highest_cases_locations = df_subset.groupby("location").agg({"new_cases": "sum"})
highest_cases_locations.sort_values("new_cases", ascending=False, inplace=True)
highest_cases_locations
```

## Top 10 Locations with highest total case

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=highest_cases_locations.head(10), x="location", y="new_cases")
plt.title("Top 10 Locations with highest total cases")
plt.ylabel("total_cases")
plt.xticks(rotation=20)
plt.show()
```

we see that US and China can compared with continents

## world wide corona cases over time

```python
df["date"] = pd.to_datetime(df["date"])
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day
```

```python
world = df[df["location"] == "World"]
sns.lineplot(data=world, x="year", y="total_cases")
plt.title("worldwide coronavirus total cases")
plt.show()
```

```python

world = df[df["location"] == "World"]
sns.lineplot(data=world, x="year", y="total_deaths")
plt.title("worldwide coronavirus total deaths")
plt.show()
```

```python
world = df[df["location"] == "World"]
sns.lineplot(data=world, x="year", y="new_cases")
plt.title("worldwide coronavirus new cases")
plt.show()
```

```python
world = df[df["location"] == "World"]
sns.lineplot(data=world, x="year", y="new_deaths")
plt.title("worldwide coronavirus new deaths")
plt.show()
```

as shown in the figures number of new cases and number of new deaths decreased in 2022 and 2023

## country specific graphs

```python
def visualise_country_cases(country):
    country_df = df[ df["location"] == country ]
    sns.lineplot(data=country_df, x="year", y="total_cases")
    plt.title(f"{country} total cases")
    plt.show()

countries = ["Egypt", "United States", "China", "Italy", "France"]
```

```python
for country in countries:
    visualise_country_cases(country)
```

```python
def visualise_country_deaths(country):
    country_df = df[df["location"] == country]
    sns.lineplot(data=country_df, x="year", y="total_deaths")
    plt.title(f"{country} total deaths")
    plt.show()
```

```python
for country in countries:
    visualise_country_deaths(country)
```

```python
plt.figure(figsize=(12, 8))
for country in countries:
    country_df = df[df["location"] == country]
    sns.lineplot(data=country_df, x="year", y="total_cases", errorbar=None)

plt.title("coronavirus total cases")
plt.legend(countries, prop={"size": 15})
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for country in countries:
    country_df = df[df["location"] == country]
    sns.lineplot(data=country_df, x="year", y="total_deaths", errorbar=None)
    
plt.title("coronavirus total deaths")
plt.legend(countries, prop={"size": 15})
plt.show()
```
